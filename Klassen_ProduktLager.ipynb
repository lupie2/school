```python
class Produkt:
    def __init__(self, name, produkt_id, preis, bestand):
        self.name = name
        self.produkt_id = produkt_id
        self.preis = preis
        self.bestand = bestand

    def anzeigen_details(self):
        print(f"Produkt: {self.name} (ID: {self.produkt_id})")
        print(f"Preis: {self.preis:.2f} EUR")
        print(f"Verfügbar: {self.bestand} Stück")

    def produkt_verkaufen(self, menge):
        if menge <= self.bestand:
            self.bestand -= menge
            print(f"{menge} Stück von {self.name} verkauft.")
        else:
            print(f"Nicht genügend {self.name} auf Lager, verfügbar: {self.bestand}, versucht zu verkaufen: {menge}")

    def bestand_auffuellen(self, menge):
        self.bestand += menge
        print(f"{menge} Stück von {self.name} dem Lager hinzugefügt.")


# Erstellen von Produkten
laptop = Produkt("Lenovo ThinkPad", "LP001", 799.00, 10)
maus = Produkt("Sharkoon Skiller SGM50W ", "MS001", 49.99, 10)
tastatur = Produkt("Logitech G213", "TT001", 74.99, 10)

# Zeige Details der Produkte an
print(laptop.anzeigen_details())
print(maus.anzeigen_details())
print(tastatur.anzeigen_details())

# Verkauf eines Produkts
laptop.produkt_verkaufen(1)

# Zu viel verkaufen mit Fehlermeldung
maus.produkt_verkaufen(11)

# Bestand auffüllen
tastatur.bestand_auffuellen(15)

# Test ob Verkauf und Auffüllen funktioniert hat
print(laptop.anzeigen_details())
print(maus.anzeigen_details())
print(tastatur.anzeigen_details())
```
