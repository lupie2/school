```python
''' WORK IN PROGRESS
working on a program that asks the user for a string and then searches the first 1000000 digits of pi for that string

1. clean pi.txt to delete spaces and new lines
2. input asking user for string
3. search pi
'''
import tkinter as tk
import re
```

```python
#the file  containing the first 1000000 digits of pi was formatted using a lot of spaces and \n commands, so I had to clean it up
file = "pi.txt"
with open(file, "r") as f:
    new = f.read()
# remove spaces and \n
new = new.replace(" ", "").replace("\n", "")

#save the modified content
with open(file, "w") as f:
    f.write(new)
```

```python
#create window with entry widget, a button and a label

root = tk.Tk()

# setting the windows size
root.geometry("600x400")

# declaring the searchable string variable
entry = tk.Entry(root)

# dictionary with conversion table to convert letters into numbers
table = {'a': '1', 'b': '2', 'c': '3', 'd': '4', 'e': '5', 'f': '6', 'g': '7', 'h': '8',
     'i': '9', 'j': '10', 'k': '11', 'l': '12', 'm': '13', 'n': '14',
     'o': '15', 'p': '16', 'q': '17', 'r': '18', 's': '19', 't': '20',
     'u': '21', 'v': '22', 'w': '23', 'x': '24', 'y': '25', 'z': '26'}

# going through the digits of pi trying to find the number
#def find():
    

# convert the string containing letters into a string of corresponding numbers
def convert(x):
    number = ''
    for i in x.lower():
        number += table[i]
    convert_label.config(text = number)
    count_label.config(text = len(number))

            
# defining the function that searches the string in pi
def search():
    sentence = entry.get().replace(" ", "")
    if not sentence.isalpha():
        result_label.config(text = "please use only English letters")
        return
    convert(sentence)
    #result_label.config(text = sentence)
    return

# entry field for users
entry.grid(row=0,column=1)

# creating labels for the results
string_text = tk.Label(root, text = 'Put in the string \nyou want to find')
string_text.grid(row=0,column=0)
result_label = tk.Label(root, text = '')
result_label.grid(row=3,column=1)
convert_text = tk.Label(root, text = 'converted string in numbers')
convert_text.grid(row=4,column=0)
convert_label = tk.Label(root, text = '')
convert_label.grid(row=4,column=1)
count_text = tk.Label(root, text = 'number of digits in given string')
count_text.grid(row=5,column=0)
count_label = tk.Label(root, text = '')
count_label.grid(row=5,column=1)
found_text = tk.Label(root, text = 'longest series of digits found in pi')
found_text.grid(row=6,column=0)

#to do
#label to show the position of found digits
#label to show probability of finding the string in pi?

# creating a button using the widget
search_btn=tk.Button(root,text = 'Search', command = search)
search_btn.grid(row=2,column=1)




root.mainloop()
```
