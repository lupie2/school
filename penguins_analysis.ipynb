```python
import pandas as pd
```

```python
# Enlesen aus Seaborn
import seaborn as sns
penguins = sns.load_dataset("penguins")
# Einlesen der csv-Datei
penguins = pd.read_csv("penguins.csv")
# Anzeigen der ersten 5 Zeilen
penguins.head()
```

```python
penguins.info()
```

```python
penguins.isnull().sum()
```

```python
### Komplett leere (Daten)zeilen löschen ### 
print(penguins.loc[penguins["bill_length_mm"].isnull()])       # Komplette Zeilen mit null in Spalte bill_lengt_mm
print(penguins.loc[penguins["bill_length_mm"].isnull()].index) # Index der Zeilen
penguins.drop(index = penguins.loc[penguins["bill_length_mm"].isnull()].index, inplace = True) # Zeilen im original Dataframe löschen
penguins
```

```python
penguins.describe().T
```

```python
### Index neu setzen (inplace=True)
### alten Index nicht in den DF schreiben (drop = True)
penguins.reset_index(drop= True, inplace=True)
penguins
```
