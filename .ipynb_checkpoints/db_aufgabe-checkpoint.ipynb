Ziel: Das Programm soll 3 Zahlen vom Benutzer einlesen, den Durchschnitt berechnen und in eine Datei schreiben.

```
import math ## unnötiger Import
```

```
def berechne_durchschnitt(zahlen):
    summe = 0
    for i in range(3): ##fehlender Doppelpunkt
        summe += zahlen[i]
    durchschnitt = summe / len(zahlen) ##Klammer war nicht geschlossen
    return durchschnitt
```

```
zahlen = []
```

```
for i in range(3):
    eingabe = input("Gib eine Zahl ein: ")
    zahlen.append(int(eingabe))
```

```
durchschnitt = berechne_durchschnitt(zahlen) ##berechne_durchschnitt war falsch geschrieben
print("Der Durchschnitt ist: " + str(durchschnitt)) ## print kann nur str concatenieren deshalb str(float)
```

```
output = open("ausgabe.txt", "w") ##Klammer war nicht geschlossen
output.write("Durchschnitt: " + str(durchschnitt))
output.close() ##close war falsch geschrieben
```

```
if durchschnitt > 10:
    print("Ziemlich hoch!") ## Klammer war nicht geschlossen
elif durchschnitt < 0:
    print("Negativer Wert!")
```
