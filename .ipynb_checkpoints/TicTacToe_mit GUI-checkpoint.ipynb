Import der notwendigen Bibliotheken

```
import tkinter as tk
from tkinter import messagebox
from random import choice
```

```
# Definition der Gewinnkombinationen
winning_combinations = [
        [(0, 0), (0, 1), (0, 2)],  # Erste Reihe
        [(1, 0), (1, 1), (1, 2)],  # Zweite Reihe
        [(2, 0), (2, 1), (2, 2)],  # Dritte Reihe
        [(0, 0), (1, 0), (2, 0)],  # Erste Spalte
        [(0, 1), (1, 1), (2, 1)],  # Zweite Spalte
        [(0, 2), (1, 2), (2, 2)],  # Dritte Spalte
        [(0, 0), (1, 1), (2, 2)],  # Diagonale von links oben nach rechts unten
        [(0, 2), (1, 1), (2, 0)]   # Diagonale von rechts oben nach links unten
]
```

```
# √úberpr√ºfung, ob alle Positionen den gleichen Wert haben und nicht leer sind - > vorhandene Gewinnkombination
def check_winner(board):
    for combo in winning_combinations:
        a, b, c = combo
        if board[a[0]][a[1]] == board[b[0]][b[1]] == board[c[0]][c[1]] != '':
            return board[a[0]][a[1]]  # Gibt 'X' oder 'O' zur√ºck, wenn eine Gewinnkombination vorliegt
    if all(cell in ['X', 'O'] for row in board for cell in row): # wenn alle Felder belegt: Unentschieden
        return 'Draw'
    return None
```

```
# Funktion zur Beendigung des Spiels, Deaktivierung der Buttons, Ausgabe der Nachrichten
def end_game(winner):
    if winner == 'Draw':
        messagebox.showinfo("Unentschieden", "Alle Felder sind belegt. Das Spiel endet unentschieden!")
    elif winner == 'O':
        messagebox.showinfo("Spiel beendet", "Herzlichen Gl√ºckwunsch! Du hast gewonnen!")
    elif winner == 'X':
        messagebox.showinfo("Spiel beendet", "Epic fail! Der Computer hat gewonnen!")

    for row in buttons:                             # Spielende -> Buttons werden disabled
        for btn in row:
            btn.config(state='disabled')
```

```
# Erstellung einer Funktion, die den Zug des Benutzers verarbeitet und die Gewinnpr√ºfung anst√∂√üt. Zug per Button-Klick.
def user_move(r, c):
    if board[r][c] == '':                                   # Pr√ºfung: gew√§hltes Feld leer?
        board[r][c] = 'O'                                   # Markierung des Feldes
        buttons[r][c].config(text='O', state='disabled')    # Button wird disabled, da Feld nun belegt
        winner = check_winner(board)                        # Durchf√ºhrung der Gewinnpr√ºfungs-Fkt.
        if winner:                                          # wenn Gewinner -> Durchf√ºhrung der Endgame-Fkt.
            end_game(winner)
        else:                                               # wenn kein Gewinner - Durchf√ºhrung der computer_move-Fkt. nach 200 ms
            root.after(200, computer_move)
```

```
# Funktion f√ºr den zuf√§lligen Zug des Computers
def computer_move():
    available_move = [(r, c) for r in range(3) for c in range(3) if board[r][c] == '']  # Pr√ºfung auf m√∂gliche Z√ºge (leere Felder)
    if available_move:
        r, c = choice (available_move)                                                  # Auswahl eines Feldes per Zufall (choice)
        board[r][c] = 'X'                                                               # Markierung des Feldes
        buttons[r][c].config(text='X', state='disabled')                                # Button wird disabled, da Feld nun belegt
        winner = check_winner(board)                                                    # Durchf√ºhrung der Gewinnpr√ºfung
        if winner:                                                                      # wenn Gewinner -> Durchf√ºhrung der Endgame-Fkt.
            end_game(winner)    
```

```
# Zur√ºcksetzen des Spiels
def reset_game():                                           # zur√ºcksetzen des Boards auf den Anfangsstatus f√ºr einen Neustart des Spiels
    global board
    board = [['' for _ in range(3)] for _ in range(3)]      # Leeren der Felder
    for r in range(3):                                      # Zur√ºcksetzen der Buttons auf "normalen" Status
        for c in range(3):
            buttons[r][c].config(text='', state='normal')
```

```
# Initalisierung der GUI
root = tk.Tk()
root.title("Tic Tac Toe mit GUI")
```

```
board = [['' for _ in range(3)] for _ in range(3)]          # 3 x 3 leeres Spielfeld
buttons = []                                                # Button-Referenzen f√ºr jedes Feld
```

```
# Erstellung des Spielboards: 3 x 3 Spielfeld mit Buttons
for r in range(3):
    row = []
    for c in range(3):
        btn = tk.Button(root, text='', width=8, height=4, font=("Arial", 24),       # Erstellung eines Buttons ohne Text mit diesen Eigenschaften
                        command=lambda r=r, c=c: user_move(r, c))                   # Anweisung der Funktion beim Klicken des Buttons
        btn.grid(row=r, column=c, padx=5, pady=5)                                   # Platzierung des Buttons
        row.append(btn)                                                             # F√ºge Button der Zeile hinzu
    buttons.append(row)                                                             # F√ºge Zeile der Button-Liste hinzu
```

```
# Erstellung eines Neustart-Buttons
reset_button = tk.Button(root, text="üîÑ Neustart", font=("Arial", 14), command=reset_game) # Erstellung eines Buttons ohne Text mit diesen Eigenschaften,
                                                                                            # Anbindung an reset_game-Fkt.
reset_button.grid(row=3, column=0, columnspan=3, pady=10)                                   # Definition der Position des Buttons
```

```
root.mainloop()
```
