**Achtung!:** Die Summe des Kassenzettels stimmt nicht.

```python
# Programm zur Darstellung des Kassenzettels

#importiert numpy für die Formatierungs-Funktion
import numpy as np

#Angabe diverser String-Variablen
receipt = 'RECEIPT'
customer = 'CUSTOMER COPY'
sternchen = '*'
sep = '\n'
sub = 'SUB TOTAL'
tax = 'SALES TAX'
equal = '========='
amount = 'AMOUNT DUE'

# Speichern der Beträge als Float in einem Array a
a = np.array([125.50, 45.00, 80.25, 12.50, 130.00, 29.95, 34.75])

# Formatierung der Beträge von Float zu String in einem Array b, damit sie mit ljust/rjust/center verwendbar sind
b = ["%.2f" % i for i in a]

print(b)

# Rechnung des tatsächlichen Betrags gespeichert als pf
pf = (a[0] + a[1] + a[2] + a[3] + a[3] + a[4] + a[5] + a[6] - a[3])

# tatsächlicher Betrag formatiert zu String als price
price = format(pf, '.2f')

# Ausrechnen der Steuersumme nach New York City Sales Tax von 8.875% gespeichert und Formatierung zu String als taxs
taxs = format((pf * 0.08875), '.2f')

# Ausrechen des Gesamtbetrags und Formatierung zu String als amountstr
amountstr = format(pf + (pf * 0.08875), '.2f')

# Überschriften des Kassenzettels
print(receipt.center(20))
print(customer.center(20))
print(20 * sternchen)

# einzeln gelistete Beträge
print(b[0].rjust(18) + ' +')
print(b[1].rjust(18) + ' +')
print(b[2].rjust(18) + ' +')
print(b[3].rjust(18) + ' +')
print(b[3].rjust(18) + ' +')
print(b[4].rjust(18) + ' +')
print(b[5].rjust(18) + ' +')
print(b[6].rjust(18) + ' +')
print(b[3].rjust(18) + ' -')

# Bisherer Betrag, Sales Tax und finaler Gesamtbetrag
print(sub + price.rjust(11))
print(tax + taxs.rjust(11))
print(equal.rjust(20))
print(amount + amountstr.rjust(10))
print(sep)
print(20 * sternchen)
```
