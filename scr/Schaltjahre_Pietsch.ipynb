```python
#Entscheidungstabelle Schaltjahre
schaltjahre = []
testj = [1720, 1972, 1980, 1600, 2000] #Jahre zum Test welche Schaltjahre sein sollen
testn = [1700, 1800, 1900, 2100] #Jahre zum Test welche keine Schaltjahre sein sollen
for i in range(1500, 2109):
    if i % 4 == 0:   #Bedingung 1: Jahreszahl ganzzahlig durch 4 teilbar
        if i % 100 == 0:    #Bedingung 2: Ausnahmen sind Zahlen die ganzzahlig durch 100 teilbar sind
            if i % 400 == 0:    #Bedingung 3: Zahlen die durch 100 UND 400 teilbar sind sind doch Schaltjahre
                schaltjahre.append(i) 
        else:
            schaltjahre.append(i)


for i in testj: #Testet ob die Jahreszahlen der Liste testj Schaltjahre sind
    if i in schaltjahre:
        print('yay')
    else:
        print('nay')
        
for i in testn: #Testet ob die Jahreszahlen der Liste testn keine Schaltjahre sind
    if i in schaltjahre:
        print('nay')
    else:
        print('yay')
        
print(schaltjahre)

```
