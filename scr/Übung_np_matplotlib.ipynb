## Kreis zeichnen

Berechnen Sie mithilfe der Funktionen np.sin() und np.cos() einen Kreis mit dem Radius r und stellen diesen mit matplotlib dar.

Denken Sie an an den Dokstring und die Dokumentation im Programm

https://numpy.org/doc/stable/reference/generated/numpy.linspace.html

```python
### schreiben Sie Ihren Code hier ###
import numpy as np
import matplotlib.pyplot as plt
def kreis(r):
    angle = np.linspace(0, 2*np.pi, 100) #berechnet 100 Werte eines Arrays die gleichmäßig im Bereich 0 bis 2*pi (voller Kreis in Bogenmaß) verteilt sind, linspace erzeugt lineare Abfolge von werten
    x = np.cos(angle)*r #berechnet die x Koordinaten des Kreises mit jedem Wert des angle-Arrays und dem Kosinus des Winkels
    y = np.sin(angle)*r #berechnet die y Koordinaten des Kreises mit jedem Wert des angle-Arrays und dem Sinus des Winkels
    plt.figure(figsize=(3, 3)) #neuer Graph mit Fenstergröße 3x3 Inches
    plt.plot(x, y) #zeichnet Kreis mit den oben berechneten x- und y-Koordinaten
### Hauptprogramm ###
#####################
kreis(1)
```

## Kreis mit Eingabe und Skalierung

Fangen Sie eventuelle Eingabefehler und Radiuswerte über 10 ab.  
Die Darstellung verschiedener Radien soll sich auch in der Ausgabe des Kreises wiederfinden.  
Beim Radius von 10 soll ca. die ganze Bildschirmbreite verwendet werden. (verwenden Sie bitte ihre Bildschirmbreite)  


```python
### schreiben Sie Ihren Code hier ###
import numpy as np
import matplotlib.pyplot as plt

def abfrage():
    while True:  # wiederhole Abfrage bis eine gültige Eingabe erfolgt
        r = input('Welchen Radius zwischen 1 und 10 soll der Kreis haben? ')
        if r.isdigit() and 1 <= int(r) <= 10:  # überprüfe ob Wert eine int zwischen 1 und 10 ist
            return int(r)  # Gibt den Wert von r zurück, wenn die Eingabe gültig ist
        else:
            print("Ungültige Eingabe! Bitte geben Sie eine Zahl zwischen 1 und 10 ein.")

def kreis(r): #Funktion aus der obigen Aufgabe
    angle = np.linspace(0, 2*np.pi, 100) #berechnet 100 Werte eines Arrays die gleichmäßig im Bereich 0 bis 2*pi (voller Kreis in Bogenmaß) verteilt sind, linspace erzeugt lineare Abfolge von werten
    x = np.cos(angle)*r #berechnet die x Koordinaten des Kreises mit jedem Wert des angle-Arrays und dem Kosinus des Winkels
    y = np.sin(angle)*r #berechnet die y Koordinaten des Kreises mit jedem Wert des angle-Arrays und dem Sinus des Winkels
    plt.figure(figsize=(r*1.2, r*1.2)) #neuer Graph dessen Größe an den Radius angepasst wird, Multiplikation mit 1.2 damit bei 10 mein Bildschirm gefüllt ist
    plt.plot(x, y) #zeichnet Kreis mit den oben berechneten x- und y-Koordinaten
    plt.plot([0, r], [0, 0], label='Radius') #zeichnet den Radius als Linie in die Grafig ein
### Hauptprogramm ###
#####################
r = abfrage()
kreis(r)
```
