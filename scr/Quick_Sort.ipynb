## als Funktion


https://de.wikipedia.org/wiki/Quicksort

Quicksort funktioniert nach dem "Teile-und-herrsche"-Prinzip ("divide and conquer"):

Als erstes teilen wir die zu sortierenden Elemente auf zwei Bereiche auf – einen mit kleinen Elementen und einen mit großen Elementen.

Welche Elemente klein sind und welche groß, entscheidet dabei das sogenannte Pivot-Element. Das Pivot-Element kann ein beliebiges Element aus dem Eingabe-Array sein. (Welches man wählt, bestimmt die Pivot-Strategie.)

Das Array wird nun so umsortiert, dass

- die Elemente, die kleiner als das Pivot-Element sind, im linken Bereich landen,
- die Elemente, die größer als das Pivot-Element sind, im rechten Bereich landen,
- dass das Pivot-Element zwischen den zwei Bereichen positioniert wird – und damit automatisch an seiner endgültigen Position.
