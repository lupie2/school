```python
#2.5 Dependent and independent events
#You are now familiar with multi-stage random experiments in which all levels in the tree diagram
#look exactly the same. However, this is not always the case. Take a look at the following
#random experiment, for example:
#A student is randomly selected from a class. There are 10 boys and 15 girls in the class.
#Of the boys, 8 study Latin and 2 study French.
#Of the girls, 3 study Latin and 12 study French.
#If you select a boy, it is much more likely that he studies Latin than if you select a
#girl. This is because 8 out of 10 boys study Latin (i.e., 80%), while only
#only 3 out of 15 (that's only 20%).
#So how likely it is that a student will learn Latin depends on whether you choose a boy or
#a girl. The two events are therefore dependent.
#If there is no connection, you call them independent . This is the case, for example, when rolling a die twice
#.
#Task:
#2
#Latin French
#8 3 12
#Latin French
#15
#Boys Girls
#10
#Calculate the probabilities for:
#➢ Grade - Boys who study Latin
#➢ Grade - Girls who study Latin
#➢ Grade - Students who study Latin
#Write a program in Python to confirm these calculations. Remember how
##the number of runs affects the accuracy of the result.

import numpy as np

iteration = int(input('number of iterations?'))
boys = 0
girls = 0
ml = 0
mf = 0
fl = 0
ff = 0
print(type(iteration))
gender = ['m', 'f']
subject = ['l', 'f']

for i in range(iteration):
    x = np.random.choice(gender, p = [0.4, 0.6])
    if x == 'm':
        boys += 1
        y = np.random.choice(subject, p = [0.8, 0.2])
        if y == 'l':
            ml += 1
        else:
            mf += 1
    else:
        girls += 1
        y = np.random.choice(subject, p = [0.2, 0.8])
        if y == 'l':
            fl += 1
        else:
            ff += 1

print('Total Boys:', boys, '    ', (boys / (boys + girls) * 100.00),'%')
print('Latin Boys:', ml, '    ', (ml / (boys + girls) * 100.00),'%', '    ', (ml / (ml + mf) * 100),'%')
print('French Boys:', mf, '    ', (mf / (boys + girls) * 100.00),'%', '    ', (mf / (ml + mf) * 100),'%')
print(100 * '#')
print('Total Girls:', girls, '    ', (girls / (boys + girls) * 100.00))
print('Latin Girls:', fl, '    ', (fl / (boys + girls) * 100.00),'%', '    ', (fl / (fl + ff) * 100),'%')
print('French Girls:', ff, '    ', (ff / (boys + girls) * 100.00),'%', '    ', (ff / (fl + ff) * 100),'%')
print(100 * '#')
print('Latin Total:', (ml + fl), '    ', ((ml+fl) / (boys + girls) * 100.00),'%')
print('French Total:', (mf + ff), '    ', ((mf+ff) / (boys + girls) * 100.00),'%')
```
