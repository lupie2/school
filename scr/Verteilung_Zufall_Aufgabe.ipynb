# Verteilung Zufallsgenerator
- Lassen Sie sich, per Zufallsgenerator, eine Zufallszahl zwischen 1 und 6 generieren.
- Definieren Sie eine Variable, die die Anzahl der Durchläufe bestimmt.
- Bei jedem Durchlauf wird eine neue Zufallszahl generiert und abhängig davon ein Zähler für diese Zahl um 1 erhöht.
Nach Beendigung der Durchläufe sollen folgende Werte pro Zähler (1 bis 6) berechnet und ausgegeben werden:
- Zählerstand
- Abweichung vom Wert (Durchläufe / 6) als Zahl und in %

**Beispiel für 1.000.000 Durchläufe**
```python
Durchläufe  :	1000000
Durchschnitt:	166666.67
Zähler 1: 166799	Diff:  132.33 	%: 0.08
Zähler 2: 166675	Diff:    8.33 	%: 0.01
Zähler 3: 167345	Diff:  678.33 	%: 0.41
Zähler 4: 166866	Diff:  199.33 	%: 0.12
Zähler 5: 166073	Diff: -593.67 	%: 0.36
Zähler 6: 166243	Diff: -423.67 	%: 0.25
```

**Führen Sie das Programm mit 100, 1.000, 10.000, 100.000 und 1.000.000 Durchläufen aus und erläutern Sie die Ergebnisse.**

````python
from random import randint
for i in range(10):
    print(randint(1,6))
````

````python
from random import randint
# Geben Sie hier Ihren Programmcode ein
durchlauf = (10, 100, 1000, 10000, 100000, 1000000, 100000000) #Anzahl Durchläufe
for x in durchlauf:
    a = 0 #1
    b = 0 #2
    c = 0 #3
    d = 0 #4
    e = 0 #5
    f = 0 #6 
    for i in range(x):
        z = randint(1,6)
        match z:
            case 1:
                a += 1
            case 2:
                b += 1
            case 3:
                c += 1
            case 4:
                d += 1
            case 5:
                e += 1
            case 6:
                f += 1
    durchschnitt = round((a + b + c + d + e + f) / 6, 2)
    print('Durchläufe: ' + str(x))
    print('Durchschnitt:' + str(durchschnitt))
    print('Zähler 1: ' + str(a) + '   ' + 'Diff:' + str(round(a - durchschnitt, 2)) + '   %:' + str(round((a - durchschnitt) / durchschnitt * 100, 2)))
    print('Zähler 2: ' + str(b) + '   ' + 'Diff:' + str(round(b - durchschnitt, 2)) + '   %:' + str(round((b - durchschnitt) / durchschnitt * 100, 2)))
    print('Zähler 3: ' + str(c) + '   ' + 'Diff:' + str(round(c - durchschnitt, 2)) + '   %:' + str(round((c - durchschnitt) / durchschnitt * 100, 2)))
    print('Zähler 4: ' + str(d) + '   ' + 'Diff:' + str(round(d - durchschnitt, 2)) + '   %:' + str(round((d - durchschnitt) / durchschnitt * 100, 2)))
    print('Zähler 5: ' + str(e) + '   ' + 'Diff:' + str(round(e - durchschnitt, 2)) + '   %:' + str(round((e - durchschnitt) / durchschnitt * 100, 2)))
    print('Zähler 6: ' + str(f) + '   ' + 'Diff:' + str(round(f - durchschnitt, 2)) + '   %:' + str(round((f - durchschnitt) / durchschnitt * 100, 2)))
````
