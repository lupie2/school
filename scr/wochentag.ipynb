```
import tkinter as tk
import math
from tkinter.messagebox import showinfo
from tkinter import ttk
```

```
window = tk.Tk()
window.geometry('450x400')
window.title('Listbox')
```

```
schaltjahre = []
for i in range(1600, 2030):
    if i % 4 == 0:   #Bedingung 1: Jahreszahl ganzzahlig durch 4 teilbar
        if i % 100 == 0:    #Bedingung 2: Ausnahmen sind Zahlen die ganzzahlig durch 100 teilbar sind
            if i % 400 == 0:    #Bedingung 3: Zahlen die durch 100 UND 400 teilbar sind sind doch Schaltjahre
                schaltjahre.append(i) 
        else:
            schaltjahre.append(i)
```

```
# label
label = ttk.Label(window, text='Wähle ein Datum aus')
label.grid(row=1, column=1)
```

```
# Array für die Tage
tage = list(range(1, 32))
list_tage = tk.Variable(value=tage)
# Label für Tage
label_t = ttk.Label(window, text='Tag')
label_t.grid(row=2, column=1)
# Listbox mit den Tagen
listbox_t = tk.Listbox(window, listvariable=list_tage, height=6, selectmode=tk.SINGLE, exportselection=False)
listbox_t.grid(row=3, column=1)
# Scrollbar für die Listbox mit Tagen
vsb_t = ttk.Scrollbar(window, orient=tk.VERTICAL, command=listbox_t.yview)
listbox_t['yscrollcommand'] = vsb_t.set
vsb_t.grid(row=3, column=2, sticky='ns')
```

```
# Array für die Monate
monate = list(range(1, 13))
list_monate = tk.Variable(value=monate)
# Label für Monate
label_m = ttk.Label(window, text='Monat')
label_m.grid(row=2, column=3)
# Listbox mit den Monaten
listbox_m = tk.Listbox(window, listvariable=list_monate, height=6, selectmode=tk.SINGLE, exportselection=False)
listbox_m.grid(row=3, column=3)
# Scrollbar für die Listbox mit Monaten
vsb_m = ttk.Scrollbar(window, orient=tk.VERTICAL, command=listbox_m.yview)
listbox_m['yscrollcommand'] = vsb_m.set
vsb_m.grid(row=3, column=4, sticky='ns')
```

```
# Array für die Jahre
jahre = list(range(1600, 2030))
list_jahre = tk.Variable(value=jahre)
# Label für Jahre
label_j = ttk.Label(window, text='Jahre')
label_j.grid(row=2, column=5)
# Listbox mit den Jahren
listbox_j = tk.Listbox(window, listvariable=list_jahre, height=6, selectmode=tk.SINGLE, exportselection=False)
listbox_j.grid(row=3, column=5)
# Scrollbar für die Listbox mit Jahren
vsb_j = ttk.Scrollbar(window, orient=tk.VERTICAL, command=listbox_j.yview)
listbox_j['yscrollcommand'] = vsb_j.set
vsb_j.grid(row=3, column=6, sticky='ns')
```

```
# Label
label_d = ttk.Label(window, text='Gewähltes Datum')
label_d.grid(row=4, column=1)
# Label wo ausgewähltes Datum angezeigt wird
datum = ttk.Label(window, text='Bitte eine Auswahl treffen')
datum.grid(row=4, column=3)
```

```
# Label
label_w = ttk.Label(window, text='Wochentag')
label_w.grid(row=5, column=1)
# Label wo ausgewähltes Datum angezeigt wird
wochentag = ttk.Label(window, text='Bitte eine Auswahl treffen')
wochentag.grid(row=5, column=3)
```

```
def daycheck():
    selected_m = listbox_m.curselection()
    selected_j = listbox_j.curselection()
    
    if selected_m and selected_j:
        monat = listbox_m.get(selected_m)
        jahr = listbox_j.get(selected_j)
        
        if monat in (4, 6, 9, 11):  # Monate mit 30 Tagen
            tage = list(range(1, 31))
        elif monat == 2:
            if jahr in schaltjahre:  # Schaltjahr
                tage = list(range(1, 30))
            else:  # Kein Schaltjahr
                tage = list(range(1, 29))
        else:  # Alle anderen Monate mit 31 Tagen
            tage = list(range(1, 32))
        
        # Tage in der Listbox aktualisieren
        list_tage.set(tage)
```

```
def handle_item_select(event):
    # Holen der ausgewählten Werte
    selected_t = listbox_t.curselection()  # Tag
    selected_m = listbox_m.curselection()  # Monat
    selected_j = listbox_j.curselection()  # Jahr
    # Wenn alle Listen eine Auswahl haben
    if selected_t and selected_m and selected_j:
        tag = listbox_t.get(selected_t)
        monat = listbox_m.get(selected_m)
        jahr = listbox_j.get(selected_j)
        #Berechnung der Kalenderwoche
        tpM = {1: 31, 2: 28, 3: 31, 4: 30, 5: 31, 6: 30, 7: 31, 8: 31, 9: 30, 10: 31, 11: 30, 12: 31}
        ordinaltag = 0
        if jahr in schaltjahre:
            tpM.update({2: 29})
        for i in range(1, monat):
            ordinaltag = ordinaltag + tpM[i]
        ordinaltag = ordinaltag + tag

        # Berechnung des Wochentags
        t = int(tag)
        m = int(monat)
        if m < 3:
            m += 12
            jahr -= 1
        y = int(str(jahr)[-2:])
        c = int(str(jahr)[:2])
        w = ('Samstag', 'Sonntag', 'Montag', 'Dienstag', 'Mittwoch', 'Donnerstag', 'Freitag')
        A = (math.floor(13 * (m + 1) / 5) + math.floor(y / 4) + math.floor(c / 4) + t + y - (2 * c) ) % 7
        if A < 0:
            A = A + 7

        # offset berechnung
        offset = 0
        t = 4
        m = 1
        if m < 3:
            m += 12
            jahr -= 1
        y = int(str(jahr)[-2:])
        c = int(str(jahr)[:2])
        w = ('Samstag', 'Sonntag', 'Montag', 'Dienstag', 'Mittwoch', 'Donnerstag', 'Freitag')
        A = (math.floor(13 * (m + 1) / 5) + math.floor(y / 4) + math.floor(c / 4) + t + y - (2 * c) ) % 7
        if A < 0:
            A = A + 7
        if A < 2:
            offset = A + 5
        else:
            offset = A - 2
        kalenderwoche = math.floor((ordinaltag - offset) / 7) + 1

        datum.config(text=f'{tag:02d}.{monat:02d}.{jahr}')
        wochentag.config(text=f'{w[A]}')
    else:
        datum.config(text='Bitte eine Auswahl treffen.')
        wochentag.config(text='Bitte eine Auswahl treffen.')
```

```
# Bindung der Listboxen mit der Auswahl-Funktion
listbox_t.bind('<<ListboxSelect>>', handle_item_select)
listbox_m.bind('<<ListboxSelect>>', lambda e: [daycheck(), daycount(), handle_item_select(e)])
listbox_j.bind('<<ListboxSelect>>', lambda e: [daycheck(), daycount(), handle_item_select(e)])
```

```
window.mainloop()
```
