# Listengenerator Zahlen
Erstellen Sie ein Programm welches eine Zufallsliste von Zahlen erzeugt.
- Es wird ein Argument für die Länge der zu erzeugenden Liste der Funktion übergeben
- Der Wertebereich der Liste ist (-2 * n) bis (4 * n)

**Beispielliste:**

[-6, 79, 29, 19, -24, 61, 69, 29, 3, 28, 69, 40, 70, -13, 31, 69, 2, 61, 25, -13]

Verwenden Sie Funktionen aus der Bibliothek 'random'
Recherchieren Sie auch andere Funktionen (Wirkungsweise, Syntax) dieser Biliothek


```python
# Zufallsliste

liste = []
def zufallsliste(n):
    import random
    for i in range(n):
        liste.append(random.randint(-2 * n, 4 * n))

### Hauptprogramm ###
zufallsliste(100)
print(liste)
```

# Listengenerator ASCII-Zeichen
Erstellen Sie ein Programm welches eine Zufallsliste von ASCII-Zeichen erzeugt.
- Es wird ein Argument für die Länge der zu erzeugenden Liste der Funktion übergeben
- Der Wertebereich der Liste sind Zeichen der ASCII-Tabelle von Positionen 33 bis 126.

**Beispielliste:**

['D','H', 'b', 'Z', 'z', 'W', 'l', '#', '~']

Verwenden Sie Funktionen aus der Bibliothek 'random'
Recherchieren Sie auch andere Funktionen (Wirkungsweise, Syntax) dieser Biliothek


```python
# Zufallsliste
import random
liste_a = []
def zufallsliste_a(n):
    for i in range(n):
        liste_a.append(chr(random.randint(33, 126)))
    
### Hauptprogramm ###
zufallsliste_a(10)
print(liste_a)
```

```python
# Zufallsliste gemischt
import random
liste_g = []
def zufallsliste_g(n):
    for i in range(n):
        if random.choice([True, False]):
            liste_g.append(random.randint(-2 * n, 4 * n))
        else:
            liste_g.append(chr(random.randint(33, 126)))
        

### Hauptprogramm ###
zufallsliste_g(10)
print(liste_g)
```
