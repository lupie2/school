```python
class Product:
    def __init__(self, name, product_id, price, stock):
        self.name = name
        self.product_id = product_id
        self.price = price
        self.stock = stock

    def product_details(self):
        print(f"Product: {self.name} (ID: {self.product_id})")
        print(f"Price: {self.price:.2f} EUR")
        print(f"In Stock: {self.stock} ")

    def product_sell(self, amount):
        if amount <= self.stock:
            self.stock -= amount
            print(f"{self.stock} items of {self.name} sold.")
        else:
            print(f"Not enough {self.name} in stock, available stock: {self.stock}, tried to sell: {amount}")

    def replenish_stock(self, amount):
        self.stock += amount
        print(f"{self.stock} items of {self.name} added to stock.")


# Generate products
laptop = Product("Lenovo ThinkPad", "LP001", 799.00, 10)
mouse = Product("Sharkoon Skiller SGM50W ", "MS001", 49.99, 10)
keyboard = Product("Logitech G213", "TT001", 74.99, 10)

# show details
print(laptop.product_details())
print(mouse.product_details())
print(keyboard.product_details())

# sell product
laptop.product_sell(1)

# sell too much with error message
mouse.product_sell(11)

# replenish stock
keyboard.replenish_stock(15)

# Testing
print(laptop.product_details())
print(mouse.product_details())
print(keyboard.product_details())
```
