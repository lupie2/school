## Blood alcohol concentration

Using the formula developed by Erik Widmark:

c = a / (m\*r)  

**c**:   blood alcohol concentration

**a**:   amount of alcohol a person has consumed, calculated by the following             formula: a = v \* e \* p  

**r**:   absorption rate (women 0.6, men 0.7)

**m**:   body weight in kg

**v**:   volume of the drink in ml

**p**:   0.8 g/ml (alcohol density)

**e**:   alcohol volume in % (e.g. beer has 4%)


We need a program that asks the user for the following information to calculate the alcohol concentration in their blood:

- alcohol content in your drink (4% for beerg, 12% for wine, etc) to assign a value to e in our formula
- volume of the drink in ml to assign it to v
- body mass in kg to assign m
- ask for gender, if female then r = 0.6, if male then r = 0.7

```python
e = float(input('What is the alcohol content of your drink in %?')) / 100
v = float(input('How much of the drink did you consume in ml?'))
m = float(input('Your weight in kg:'))
gender = input('Your gender (m/f)?')
p = 0.8 #alcohol density
r = None
a = v * e * p #amount of alcohol consumed

match gender.upper():
        case 'F':
            r = 0.6
        case 'M':
            r = 0.7
        case _:
            print('Invalid input, please enter f or m')
            geschlecht = input('Your gender (m/f)?')

c = a / (m * r)

print('The alcohol concentration in your blood is ' + str(c) + '%')
```
